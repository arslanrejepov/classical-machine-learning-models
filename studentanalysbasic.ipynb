```python
import pandas as pd
import numpy as np
```

```python
names = np.array(['Ali', 'Ayna', 'Arslan', 'Mira', 'Begench'])
ages = np.array([19, 21, 20, 22, 23])
scores = np.array([88, 92, 80, 95, 85])
```

```python
#dataframe
df=pd.DataFrame({
    'name':names,
    'age':ages,
    'scores':scores})
print(df)
```

```python
mean_score=np.mean(scores)
print("Average score",mean_score)
```

```python
high_scores=df[df['scores']>85]
print("Scores >85")
print(high_scores)
```

```python
df['passed'] = df['scores'] >= 85
print("\nTäze sütun goşulandan soň:")
print(df)
```
