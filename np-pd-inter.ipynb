```python
# Given a DataFrame with missing values, fill numeric columns with median and categorical with mode
import pandas as pd

df=pd.DataFrame({
    'Age': [25, 30, None, 22, 28, None],
    'Salary': [50000, None, 60000, 55000, None, None],
    'City': ["Hangzhou", None, None, "Shanghai", "Beijing", None]
})

for col in df.columns:
    if df[col].dtype in [int, float]:
        df[col].fillna(df[col].median(), inplace=True)
    else:
        df[col].fillna(df[col].mode()[0], inplace=True)
    
print(df)

```
